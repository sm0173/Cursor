Почему вообще используют `numpy`?

```
n = 300
A = np.random.rand(n, n)
B = np.random.rand(n, n)
```

```
%%time
C = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        for k in range(n):
            C[i, j] += A[i, k] * B[k, j]
```

```
%%time
C = A @ B
```
