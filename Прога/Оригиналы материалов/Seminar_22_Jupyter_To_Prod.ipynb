## 1. Командная строка (Bash)

### Синтаксис Bash

Ниже представлен основной список команд, его подробно разбирать не надо, это скорее как справочник, лучше сразу переходить к выполнению каких-то простых команд для интуитивного понимания работы с командной строкой

#### Навигация по файловой системе

* `pwd` — **P**rint **W**orking **D**irectory — показывает текущую директорию (где вы сейчас находитесь)

* `ls` — **L**i**S**t — показывает список файлов и папок в текущей директории
  
  `ls -l` — подробный список (права, размер, дата)
  
  `ls -a` — показать все файлы, включая скрытые (начинающиеся с `.`)
  
  `ls -la` — комбинация: подробный список всех файлов

* `cd` — **C**hange **D**irectory — перейти в другую директорию
  
  `cd folder` — перейти в папку folder
  
  `cd ..` — перейти на уровень выше (в родительскую папку)
  
  `cd ~` — перейти в домашнюю директорию
  
  `cd /` — перейти в корень файловой системы

#### Работа с файлами и папками

* `mkdir` — **M**a**K**e **DIR**ectory — создать новую папку
  
  `mkdir folder` — создать папку с именем folder
  
  `mkdir -p path/to/folder` — создать папку со всеми промежуточными директориями

* `touch` — создать пустой файл или обновить время модификации существующего
  
  `touch file.txt` — создать файл file.txt

* `cp` — **C**o**P**y — копировать файл или папку
  
  `cp source.txt dest.txt` — копировать файл
  
  `cp -r folder1 folder2` — копировать папку рекурсивно

* `mv` — **M**o**V**e — переместить или переименовать файл/папку
  
  `mv old.txt new.txt` — переименовать файл
  
  `mv file.txt folder/` — переместить файл в папку

* `rm` — **R**e**M**ove — удалить файл или папку
  
  `rm file.txt` — удалить файл
  
  `rm -r folder` — удалить папку рекурсивно

#### Просмотр содержимого файлов

* `cat` — con**CAT**enate — показать всё содержимое файла
  
  `cat file.txt` — вывести содержимое file.txt

* `head` — показать первые строки файла
  
  `head file.txt` — первые 10 строк (по умолчанию)
  
  `head -n 5 file.txt` — первые 5 строк

* `tail` — показать последние строки файла
  
  `tail file.txt` — последние 10 строк
  
  `tail -n 5 file.txt` — последние 5 строк

* `less` — постраничный просмотр файла (не работает в Google Colab)

#### Создание и запись в файлы

* `echo` — вывести текст или записать его в файл
  
  `echo "text"` — просто вывести text
  
  `echo "text" > file.txt` — записать text в файл (перезапись)
  
  `echo "text" >> file.txt` — добавить text в конец файла

* `>` — перенаправление вывода в файл (перезапись содержимого)

* `>>` — перенаправление вывода в файл (добавление в конец)

#### Поиск и фильтрация

* `grep` — **G**lobal **R**egular **E**xpression **P**rint — поиск текста по паттерну
  
  `grep "pattern" file.txt` — найти строки, содержащие pattern
  
  `grep -i "pattern" file.txt` — игнорировать регистр
  
  `grep -n "pattern" file.txt` — показать номера строк
  
  `grep -v "pattern" file.txt` — инвертировать поиск (строки БЕЗ pattern)

* `find` — поиск файлов по имени или другим критериям
  
  `find . -name "*.txt"` — найти все .txt файлы в текущей папке
  
  `find . -type f` — найти только файлы (не папки)
  
  `find . -type d` — найти только папки

* `wc` — **W**ord **C**ount — подсчёт строк, слов, символов
  
  `wc file.txt` — вывести количество строк, слов и байт
  
  `wc -l file.txt` — только количество строк
  
  `wc -w file.txt` — только количество слов

#### Комбинирование команд

* `|` — **pipe** (конвейер) — передать вывод одной команды на вход другой
  
  `cat file.txt | grep "error"` — найти строки с "error" в файле
  
  `ls -la | grep ".txt"` — найти только .txt файлы в списке

* `&&` — выполнить следующую команду, только если предыдущая успешна
  
  `mkdir folder && cd folder` — создать папку И перейти в неё

* `;` — выполнить команды последовательно (независимо от результата)
  
  `cd folder; ls` — перейти в папку; показать содержимое

#### Полезные утилиты

* `sort` — отсортировать строки
  
  `sort file.txt` — сортировать по алфавиту
  
  `sort -r file.txt` — обратная сортировка
  
  `sort -n file.txt` — числовая сортировка

* `uniq` — убрать дублирующиеся строки (работает только на отсортированных данных!)
  
  `sort file.txt | uniq` — убрать дубликаты

* `sed` — **S**tream **ED**itor — поиск и замена текста
  
  `sed 's/old/new/' file.txt` — заменить первое вхождение old на new
  
  `sed 's/old/new/g' file.txt` — заменить все вхождения
  
  `sed -i 's/old/new/g' file.txt` — заменить в самом файле (**i**n-place)

#### Переменные и окружение

* `$VARIABLE` — обращение к переменной
  
  `echo $HOME` — вывести путь к домашней директории
  
  `echo $PATH` — вывести переменную PATH

* `export` — установить переменную окружения
  
  `export MY_VAR="value"` — создать переменную MY_VAR

#### Специальные символы

* `*` — **wildcard** (шаблон) — любое количество любых символов
  
  `ls *.txt` — все файлы с расширением .txt
  
  `rm test*` — удалить все файлы, начинающиеся с test

* `?` — ровно один любой символ
  
  `ls file?.txt` — найдёт file1.txt, fileA.txt, но не file10.txt

* `~` — домашняя директория текущего пользователя
  
  `cd ~` — перейти домой

* `.` — текущая директория

* `..` — родительская директория (уровень выше)

## 2. Обучение и сохранение модели

### План

Давайте обучим модель предсказания из **16 семинара по ML basics** и сохраним её в файл.

### Зачем это делать?

Чаще всего модель обучается **один раз** (иногда это занимает часы или дни), а используется **многократно**. Сохранение модели в файл позволяет не переобучать её каждый раз при запуске приложения.

**Типичный сценарий выглядит так:**
1. Модель обучается на мощном сервере или в облаке
2. Сохраняется в файл
3. Этот файл загружается в API или веб-приложение для выполнения предсказаний

Для начала подготовим данные и создадим удобную функцию для подсчета метрик

**Рекомендуется** сразу скинуть нижний код студентам, чтобы они шли в одном темпе

````python
# Скачаем наш датасет
! wget https://www.dropbox.com/s/ofr99e7skcd515m/telecom_churn.csv
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Считаем наш датасет при помощи библиотеки pandas
data = pd.read_csv('telecom_churn.csv')

X = data.drop(columns='Churn')
y = data['Churn']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=12)

def make_scores(y_valid, preds_valid):
    # Посчитаем метрики для нашего алгоритма
    print(f'Accuracy  = {accuracy_score(y_valid, preds_valid)}')
    print(f'Precision = {precision_score(y_valid, preds_valid)}')
    print(f'Recall    = {recall_score(y_valid, preds_valid)}')
````

**Вопрос**: где скорее всего сейчас лежит **`telecom_churn.csv`**? Какой командой проверим?

````python
# перенесем сразу telecom_churn.csv в репозиторий нашего проекта для дальнейшего удобства

!mkdir /content/my_project/data && mv /content/telecom_churn.csv /content/my_project/data
````

Обучим первую модель

````python
knn_first = KNeighborsClassifier(n_neighbors=3)
knn_first.fit(X_train, y_train)
````

Теперь давайте сохраним обученную модель в [бинарный файл](https://blog.skillfactory.ru/glossary/binarnyy-fayl/), файл, содержащий данные в виде последовательности байтов (нулей и единиц), которые компьютер понимает напрямую. В отличие от **текстовых файлов** (`.txt`, `.csv`, `.json`), которые можно открыть и прочитать в обычном текстовом редакторе, бинарные файлы выглядят как набор нечитаемых символов.

Примеры бинарных файлов: изображения (`.jpg`, `.png`), видео (`.mp4`), исполняемые программы (`.exe`), архивы (`.zip`) — и наши сохранённые модели (`.pkl`).

**Преимущества бинарных файлов:**
- Компактность (занимают меньше места)
- Быстрая загрузка и запись
- Сохраняют точную структуру объекта

Существует стандартная библиотека [pickle](https://habr.com/ru/companies/otus/articles/888262/) для сохранения Python-объектов в бинарные файлы, давайте воспользуемся ею!

````python
!cd my_project && mkdir models # создадим директорию для хранения моделей
````

````python
import pickle

model_path = "/content/my_project/models/first_model.pkl" # куда кладем модель
with open(model_path, 'wb') as f:  # 'wb' = write binary (запись в бинарном режиме)
    pickle.dump(knn_first, f)
````

Скачаем нашу модель и протестируем ее

````python
with open(model_path, 'rb') as f:  # 'rb' = read binary (чтение в бинарном режиме)
    loaded_model = pickle.load(f)


preds_valid = loaded_model.predict(X_valid)
make_scores(y_valid, preds_valid)
````

## 3. Модульная структура проекта

### Почему нужна структура?

В реальных проектах весь код организован в виде отдельных Python-модулей (`.py` файлов), а не хранится в одном ноутбуке как минимум по следующим причинам:

**Масштабируемость:** когда проект растёт, один ноутбук с тысячами строк кода становится нечитаемым и неуправляемым. Разделение на модули позволяет логически структурировать код.

**Переиспользование:** функции и классы из отдельных модулей можно импортировать в разные части проекта, в других ноутбуках или скриптах. Не нужно копировать один и тот же код.

**Командная работа:** когда над проектом работает несколько человек, модульная структура позволяет разделить зоны ответственности — один занимается обработкой данных, другой моделью, третий оценкой результатов.

**Тестирование:** отдельные модули легко покрыть unit-тестами, что критично для production-кода.

---

Эти принципы применимы к любому проекту — будь то backend-разработка, frontend, DevOps или что-то ещё. Но, как вы уже догадываетесь, данная семинарская тетрадка имитирует простой пример ML-проекта.

Дальше мы создадим несколько модулей, которые будут отвечать за минимально необходимые вещи для работы pipeline:
- **Обработка и подготовка данных** для модели
- **Сама модель** (обучение, предсказание, сохранение/загрузка)
- **Непосредственный запуск** тестирование и обучение

#### Модель `model.py`

Теперь давайте создадим класс `Classifier`, который будет оборачивать нашу модель и предоставлять удобный интерфейс для работы с ней.

Класс будет уметь:
- **Инициализироваться** через переданный путь к сохранённому бинарному файлу модели
- **Обучаться** на переданных данных
- **Делать предсказания** на новых данных
- **Возвращать уверенность модели** — метод, который вычисляет среднюю вероятность предсказанного класса (этот метод напрямую не доступен у KNN из sklearn, нужно дополнительно обработать результат `predict_proba`)

Метод с уверенностью — это банальный пример того, что может понадобиться для бизнес-логики компании. Например, компания может хотеть показывать предсказания пользователю только если модель уверена больше чем на 80%, или отправлять неуверенные случаи на ручную проверку аналитикам. Именно поэтому и существуют такие интерфейсы-обёртки — через них мы добавляем кастомные методы и адаптируем модель под конкретные бизнес-требования.

````python
class Classifier:
    """
    Обёртка для модели классификации
    """

    def __init__(self, model_path):
        """
        Инициализация классификатора

        Args:
            model_path: путь к сохранённой модели
        """
        self.load_model(model_path)


    def train(self, X_train, y_train):
        """
        Обучение модели

        Args:
            X_train: признаки для обучения
            y_train: целевая переменная
        """
        self.model.fit(X_train, y_train)


    def predict(self, X):
        """
        Предсказание классов

        Args:
            X: данные для предсказания

        Returns:
            Предсказанные классы
        """
        return self.model.predict(X)


    def get_prediction_confidence(self, X):
        """
        Вычисляет среднюю уверенность модели в предсказаниях

        Args:
            X: данные для предсказания

        Returns:
            Средняя уверенность (0-1)
        """
        probabilities = self.model.predict_proba(X)
        predictions = self.predict(X)

        # Берём вероятности предсказанных классов
        confidence_scores = []
        for i, pred in enumerate(predictions):
            confidence_scores.append(probabilities[i][pred])

        return np.mean(confidence_scores)

    def save_model(self, path: str):
        """
        Сохранение модели в файл

        Args:
            path: путь для сохранения
        """
        with open(path, 'wb') as f:
            pickle.dump(self.model, f)


    def load_model(self, path: str):
        """
        Загрузка модели из файла

        Args:
            path: путь к файлу модели
        """
        with open(path, 'rb') as f:
            self.model = pickle.load(f)
````

## 4. Проблема зависимостей и роль Docker

Логично, ведь мы только что удалили этот модуль!

### Что произошло?

Мы удалили библиотеку `scikit-learn`, и скрипт перестал работать. Это иллюстрирует реальную проблему в разработке.

### Типичная ситуация

Вы клонируете репозиторий с GitHub или копируете чей-то код к себе на компьютер, в котором написано множество импортов...

Но все эти импорты — это **внешние** библиотеки (зависимости проекта), и нет никакой гарантии, что они уже установлены на вашем компьютере

---

### В Jupyter-тетрадке мы привыкли решать это так:

1. Запустили ячейку
2. Получили ошибку импорта
3. Выполнили `!pip install pandas`
4. Перезапустили ячейку
5. Работает!

### Но в production-окружении так делать нельзя:

- Нельзя вручную устанавливать библиотеки на сервере каждый раз
- Нужна гарантия, что окружение **одинаковое везде**
- Приложение должно запускаться **автоматически** без ручного вмешательства

---

### Как Docker решает эту проблему?

**Docker** при сборке контейнера делает примерно следующее:

1. Берёт чистую операционную систему (базовый образ)
2. Читает файл `requirements.txt` — список всех необходимых библиотек
3. **Автоматически устанавливает все зависимости**: `pip install -r requirements.txt`
4. Копирует ваш код в контейнер
5. Запускает приложение

### Таким образом гарантируется, что:

У всех одинаковое окружение (у вас, на сервере, у коллег)  
Все зависимости установлены **до** запуска кода  
Не нужно ничего устанавливать вручную



Давайте сымитируем одну из ключевых операций Docker — установим все зависимости проекта из файла `requirements.txt` перед запуском.

## 5. Тестирование кода

Когда речь идет о production-коде, тестирование становится критически важным. Тесты помогают:

- **Убедиться, что код работает правильно** — функции делают то, что должны
- **Предотвратить регрессии** — новые изменения не ломают существующий функционал
- **Документировать поведение** — тесты показывают, как должен работать код
- **Ускорить разработку** — можно быстро проверить, что всё работает после изменений
- **Повысить уверенность при деплое** — если тесты прошли, код готов к production



**[pytest](https://habr.com/ru/companies/beget/articles/948806/)** — самый популярный фреймворк для тестирования в Python. Он позволяет писать простые и понятные тесты.

Напишем пару примеров тестов для Classifier из `model.py`. Проверим, что он выдает предсказания с совпадающей размерностью обучающих данных по количеству семплов, а также что уверенность лежит в диапозоне (0; 1)
